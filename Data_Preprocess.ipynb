```python
# --> Import Libraries
import numpy as np
import pandas as pd
```

```python
# --> Read in per 100 posession data
per_100_poss = pd.read_csv("per100poss_historic_data.csv")
per_100_poss.head(5)
```

```python
per_100_poss.shape
```

```python
# Get Year from season
per_100_poss['Year'] = np.where(per_100_poss['Season'].str[-2:] == '00', '2000',
                              per_100_poss['Season'].str[:2] + per_100_poss['Season'].str[-2:])

# Convert this to an integer
per_100_poss['Year'] = per_100_poss['Year'].astype('int64')

per_100_poss
```

```python
# Let's merge per 100 possession data w/ All-NBA data (flags)
# Add key_id to possession data

per_100_poss["key_id"] = per_100_poss["Player"] + "-" + per_100_poss["Year"].astype(str)

per_100_poss.shape
```

```python
per_100_poss
```

```python
# Players that change team during a seaon will be recorded multiple times.
# The aggregate coding is Tm == "TOT"

per_100_poss.drop_duplicates(subset = ['key_id'], keep = 'first', inplace=True)
per_100_poss.shape
```

```python
per_100_poss
```

```python
# --> All NBA Data
all_nba_dat = pd.read_csv("historic_all_nba_selections.csv")
all_nba_dat["key_id"] = all_nba_dat["Player"] + "-" + all_nba_dat["Year"].astype(str)

# Change column name (b/c "Team" in the per 100 possesion data)
all_nba_dat.rename(columns={"Team": "All_NBA_Team"}, inplace=True)
all_nba_dat.tail(20)
```

```python
# --> Merge per 100 possession data with All-NBA data
merged_dat = pd.merge(left=per_100_poss, right=all_nba_dat[["All_NBA_Team", "key_id"]],
                      how='left', left_on='key_id', right_on='key_id')

# --> Replace all NaNs with 0s (NaN on 3P% means no 3s attempted)
merged_dat.fillna(0, inplace=True)


# Aesthetic to make legend levels more meaningful
# Note: np.where = ifelse()
merged_dat['All_NBA_Flag'] = np.where(merged_dat['All_NBA_Team']==int(0), "Not Selected",
                                np.where(merged_dat['All_NBA_Team']==int(1), "1st", 
                                np.where(merged_dat['All_NBA_Team']==int(2), "2nd", "3rd")        
                                ))

merged_dat
```

```python
# Add All-NBA Position flag
# I.e. Convert PG,SG into Guards, SF,PF into forwards and C into Centers
def all_nba_position(Postion_Column):
    guard_index = Postion_Column.astype(str).str.find("G") !=-1
    forward_index = Postion_Column.astype(str).str.find("F") !=-1
    center_index = Postion_Column.astype(str).str.find("C") !=-1
    
    
    result = np.where(forward_index, "Forward",
                      np.where(guard_index, "Guard",
                               np.where(center_index, "Center", "Error"))
                                    )
    
    return result
```

```python
merged_dat['All_NBA_Pos'] = all_nba_position(merged_dat['Pos'])
merged_dat.head(5)
```

### Add team performance

```python
team_data = pd.read_csv('historic_team_wins.csv')
team_data.head()
#team_data.shape
```

```python
team_data.fillna(value = 0, inplace = True)

team_data['Total_wins'] = team_data.loc[:,"ATL":"WAS"].sum(axis = 1)

team_data
```

82*15 = 1230
#team_data.iloc[7]

Not all seasons had 1230 wins recorded. For example, in the 2012-2013 season, both the Celtics and Pacers only played 81 games ( 1 game cancelled due to Boston bombing accident). Hence, the 2012-2013 season only saw 1229 wins. The 2011-2012 season was a lockout season (990 wins recorded).

```python
team_data_prop  = team_data.loc[:,"ATL":"WAS"].div(team_data['Total_wins'], axis=0)


#pd.set_option('display.max_columns', None)



team_data_prop['TOT'] = team_data_prop.mean(axis=1)
team_data_prop['Season'] = team_data['Season']

team_data_prop.head(10)


#Check
#team_data_prop.sum(axis = 1)
```

```python
np.sort(merged_dat.Tm.unique())
```

'SEA should change to OKC

```python
Team = np.where( np.logical_or(merged_dat['Tm'] == "CHA",merged_dat['Tm'] =="CHH"), "CHO",
        np.where(merged_dat['Tm'] == 'NJN', 'BRK',
                np.where( np.logical_or(merged_dat['Tm']== 'NOH',merged_dat['Tm']== 'NOK'), 'NOP',
                        np.where(merged_dat['Tm']== 'SEA', 'OKC',
                                np.where(merged_dat['Tm']== 'VAN', 'MEM',
                                        np.where(merged_dat['Tm']=="WSB", 'WAS', merged_dat['Tm']))))))



#np.sort(merged_dat.Tm.unique())

merged_dat['Team30'] = Team 

merged_dat
```

```python
team = np.array(merged_dat['Team30'])
season = np.array(merged_dat['Season'])

win_prop=[]
for index in range(len(team)):
    current_team = team[index]
    current_season = season[index]
    to_add = team_data_prop[team_data_prop['Season'] == current_season].loc[:,current_team].values[0]
    win_prop.append(to_add)
```

```python
merged_dat['Win_Prop'] = pd.Series(win_prop)
```
